```python
import os
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
# pagina exemplo https://www.valor.com.br/carteira-valor-iframe/historico/7/2020

file_name = '2020_7_carteira_valor.html'
mes = file_name.split('_')[1]
ano = file_name.split('_')[0]
print(ano, mes)
file_path = os.path.join('htmls', file_name)

tabelas = pd.read_html(file_path)
```

```python
# carteira valor
df_carteira_valor = tabelas[0]

# preenche as colunas ano e mês
df_carteira_valor['ano'] = ano
df_carteira_valor['mes'] = mes.zfill(2)
df_carteira_valor['anomes'] = ano+mes

# filtra apenas as colunas que eu quero
df_carteira_valor = df_carteira_valor[[
    'ano',
    'mes',
    'anomes',
    'Ações',
    'Var. %*',
    'Indicações'
]]

# renomeia as colunas
a_renomear = {
    'Ações': 'Ações',
    'Var. %*': 'Desempenho',
    'Indicações': 'Indicações'
}

df_carteira_valor.rename(a_renomear, axis='columns', inplace=True)

df_carteira_valor
```

```python
for df in tabelas:
    #print(df)
    print('-------------------------------------------------------------------------')
```

```python
# resultado carteira valor
df = tabelas[1]

df = df.transpose()

headers = df.iloc[0]
df  = pd.DataFrame(df.values[1:], columns=headers)

# preenche as colunas ano e mês
df['ano'] = ano
df['mes'] = mes.zfill(2)
df['anomes'] = ano+mes

df['nome'] = 'carteira_valor'

# renomeia as colunas
a_renomear = {
    'Média no mês': 'media_mes',
    'Acumulado no ano 1': 'acumulado_ano',
    'Acumulado em 12 meses 1': 'acumulado_12_meses'
}

df.rename(a_renomear, axis='columns', inplace=True)

# filtra apenas as colunas que eu quero
df = df[[
    'ano',
    'mes',
    'anomes',
    'nome',
    'media_mes',
    'acumulado_ano',
    'acumulado_12_meses'
]]

# divide por 100 os campos numéricos
df['media_mes'] = df['media_mes']/100
df['acumulado_ano'] = df['acumulado_ano']/100
df['acumulado_12_meses'] = df['acumulado_12_meses']/100

df
```

```python
# resultado carteira valor
df = tabelas[2]

df = df.transpose()

headers = df.iloc[0]
df  = pd.DataFrame(df.values[1:], columns=headers)

# preenche as colunas ano e mês
df['ano'] = ano
df['mes'] = mes.zfill(2)
df['anomes'] = ano+mes

df['nome'] = 'ibovespa'

df.dtypes

# renomeia as colunas
a_renomear = {
    'No mês': 'media_mes',
    'No ano 1': 'acumulado_ano',
    'Acumulado em 12 meses 1': 'acumulado_12_meses'
}

df.rename(a_renomear, axis='columns', inplace=True)

# filtra apenas as colunas que eu quero
df = df[[
    'ano',
    'mes',
    'anomes',
    'nome',
    'media_mes',
    'acumulado_ano',
    'acumulado_12_meses'
]]

# converte para número
df['media_mes'] = pd.to_numeric(df['media_mes'])
df['acumulado_ano'] = pd.to_numeric(df['acumulado_ano'])
df['acumulado_12_meses'] = pd.to_numeric(df['acumulado_12_meses'])

# divide por 100 os campos numéricos
df['media_mes'] = df['media_mes']/100
df['acumulado_ano'] = df['acumulado_ano']/100
df['acumulado_12_meses'] = df['acumulado_12_meses']/100

df
```

```python
#evolucao_carteira 3, remove ultimo registro informativo
df = tabelas[3][:-1]

# preenche as colunas ano e mês
df['ano'] = ano
df['mes'] = mes.zfill(2)
df['anomes'] = ano+mes

a_renomear = {
    'Unnamed: 1': 'corretora', 
    'Mês %': 'desempenho_mes', 
    'Ano % 2': 'desempenho_ano'
}

df.rename(a_renomear, axis='columns', inplace=True)

# filtra apenas as colunas que eu quero
df = df[[
    'ano',
    'mes',
    'anomes',
    'corretora',
    'desempenho_mes',
    'desempenho_ano'
]]

# converte para número
df['desempenho_mes'] = pd.to_numeric(df['desempenho_mes'])
df['desempenho_ano'] = pd.to_numeric(df['desempenho_ano'])

# divide por 100 os campos numéricos
df['desempenho_mes'] = df['desempenho_mes']/100
df['desempenho_ano'] = df['desempenho_ano']/100

df
```
