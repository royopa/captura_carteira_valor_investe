```python
import os
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
def converte_tabela_carteira_valor(df):

    # preenche as colunas ano e mês
    df['ano'] = ano
    df['mes'] = mes.zfill(2)
    df['ano_mes'] = ano+mes

    # filtra apenas as colunas que eu quero
    df = df[[
        'ano',
        'mes',
        'ano_mes',
        'Ações',
        'Var. %*',
        'Indicações'
    ]]

    # renomeia as colunas
    a_renomear = {
        'Ações': 'acao',
        'Var. %*': 'desempenho',
        'Indicações': 'indicacoes'
    }

    df.rename(a_renomear, axis='columns', inplace=True)

    df

    # converte para número
    df['desempenho'] = pd.to_numeric(df['desempenho'])
    df['indicacoes'] = pd.to_numeric(df['indicacoes'])

    # divide por 100 os campos numéricos
    df['desempenho'] = df['desempenho']/100

    return df
```

```python
def converte_tabela_resultado_carteira(df):
    df = df.transpose()

    headers = df.iloc[0]
    df  = pd.DataFrame(df.values[1:], columns=headers)

    # preenche as colunas ano e mês
    df['ano'] = ano
    df['mes'] = mes.zfill(2)
    df['ano_mes'] = ano+mes

    df['nome'] = 'carteira_valor'

    # renomeia as colunas
    a_renomear = {
        'Média no mês': 'media_mes',
        'Acumulado no ano 1': 'acumulado_ano',
        'Acumulado em 12 meses 1': 'acumulado_12_meses'
    }

    df.rename(a_renomear, axis='columns', inplace=True)

    # filtra apenas as colunas que eu quero
    df = df[[
        'ano',
        'mes',
        'ano_mes',
        'nome',
        'media_mes',
        'acumulado_ano',
        'acumulado_12_meses'
    ]]

    # converte para número
    df['media_mes'] = pd.to_numeric(df['media_mes'])
    df['acumulado_ano'] = pd.to_numeric(df['acumulado_ano'])
    df['acumulado_12_meses'] = pd.to_numeric(df['acumulado_12_meses'])

    # divide por 100 os campos numéricos
    df['media_mes'] = df['media_mes']/100
    df['acumulado_ano'] = df['acumulado_ano']/100
    df['acumulado_12_meses'] = df['acumulado_12_meses']/100

    return df
```

```python
def converte_tabela_resultado_ibovespa(df):
    df = df.transpose()

    headers = df.iloc[0]
    df  = pd.DataFrame(df.values[1:], columns=headers)

    # preenche as colunas ano e mês
    df['ano'] = ano
    df['mes'] = mes.zfill(2)
    df['ano_mes'] = ano+mes

    df['nome'] = 'ibovespa'

    df.dtypes

    # renomeia as colunas
    a_renomear = {
        'No mês': 'media_mes',
        'No ano 1': 'acumulado_ano',
        'Acumulado em 12 meses 1': 'acumulado_12_meses'
    }

    df.rename(a_renomear, axis='columns', inplace=True)

    # filtra apenas as colunas que eu quero
    df = df[[
        'ano',
        'mes',
        'ano_mes',
        'nome',
        'media_mes',
        'acumulado_ano',
        'acumulado_12_meses'
    ]]

    # converte para número
    df['media_mes'] = pd.to_numeric(df['media_mes'])
    df['acumulado_ano'] = pd.to_numeric(df['acumulado_ano'])
    df['acumulado_12_meses'] = pd.to_numeric(df['acumulado_12_meses'])

    # divide por 100 os campos numéricos
    df['media_mes'] = df['media_mes']/100
    df['acumulado_ano'] = df['acumulado_ano']/100
    df['acumulado_12_meses'] = df['acumulado_12_meses']/100

    return df
```

```python
def converte_tabela_evolucao(df):
    # preenche as colunas ano e mês
    df['ano'] = ano
    df['mes'] = mes.zfill(2)
    df['ano_mes'] = ano+mes

    a_renomear = {
        'Unnamed: 1': 'corretora', 
        'Mês %': 'desempenho_mes', 
        'Ano % 2': 'desempenho_ano'
    }

    df.rename(a_renomear, axis='columns', inplace=True)

    # filtra apenas as colunas que eu quero
    df = df[[
        'ano',
        'mes',
        'ano_mes',
        'corretora',
        'desempenho_mes',
        'desempenho_ano'
    ]]

    # converte para número
    df['desempenho_mes'] = pd.to_numeric(df['desempenho_mes'])
    df['desempenho_ano'] = pd.to_numeric(df['desempenho_ano'])

    # divide por 100 os campos numéricos
    df['desempenho_mes'] = df['desempenho_mes']/100
    df['desempenho_ano'] = df['desempenho_ano']/100

    return df
```

```python
def converte_tabela_corretora(df):
    # preenche as colunas ano e mês
    df['ano'] = ano
    df['mes'] = mes.zfill(2)
    df['ano_mes'] = ano+mes

    # pega o nome da corretora
    nome_corretora = df.columns.values[0]

    #insere a coluna com o nome da corretora
    df['corretora'] = nome_corretora

    # renomeia as colunas
    a_renomear = {
        df.columns.values[0]: 'acao', 
        'Variação - em %': 'variacao',
    }

    df.rename(a_renomear, axis='columns', inplace=True)

    # filtra apenas as colunas que eu quero

    df = df[[
        'ano',
        'mes',
        'ano_mes',
        'corretora',
        'acao',
        'variacao'
    ]]

    # converte para número
    df['variacao'] = pd.to_numeric(df['variacao'])

    # divide por 100 os campos numéricos
    df['variacao'] = df['variacao']/100

    return df
```

```python
def converte_tabela_corretora_o_que_mudou(df):
    # preenche as colunas ano e mês
    df['ano'] = ano
    df['mes'] = mes.zfill(2)
    df['ano_mes'] = ano+mes

    # pega o nome da corretora
    nome_corretora = df.columns.values[0]
    
    #insere a coluna com o nome da corretora
    df['corretora'] = nome_corretora

    # renomeia as colunas
    a_renomear = {
        df.columns.values[0]: 'descricao'
    }

    df.rename(a_renomear, axis='columns', inplace=True)

    # filtra apenas as colunas que eu quero
    df = df[[
        'ano',
        'mes',
        'ano_mes',
        'corretora',
        'descricao'
    ]]
    
    # remove espaços adicionais na descrição
    df['descricao'] = df['descricao'].str.strip()
    
    return df
```

```python
# percorre todos os arquivos
for file_name in sorted(os.listdir('htmls')):
    if file_name.endswith('.html') is False:
        continue

    mes = file_name.split('_')[1]
    mes = mes.zfill(2)
    ano = file_name.split('_')[0]

    print('='*50)
    print('Processando', file_name, ano, mes)
    print('='*50)
    
    file_path = os.path.join('htmls', file_name)
    tabelas = pd.read_html(file_path)

    print(' Carteira Valor - Início')
    # carteira valor
    df = converte_tabela_carteira_valor(tabelas[0])
    # salvando arquivo
    save_header = True
    saida_csv = 'carteira_valor.csv'
    saida_csv = os.path.join('dados', saida_csv)
    if os.path.exists(saida_csv):
        save_header = False
    df.to_csv(saida_csv, mode='a', header=save_header, index=False)
    print('   Salvando arquivo', saida_csv, 'Ok')

    
    print(' Resultado carteira valor e ibovespa - Início')
    # resultado carteira valor
    df = converte_tabela_resultado_carteira(tabelas[1])
    # resultado carteira ibovespa
    df2 = converte_tabela_resultado_ibovespa(tabelas[2])
    # junta as duas num único dataframe, para ser importado
    # numa única tabela
    df = pd.concat([df,df2.reindex(columns=df.columns)])    
    
    # salvando arquivo
    save_header = True
    saida_csv = 'resultado_carteira.csv'
    saida_csv = os.path.join('dados', saida_csv)
    if os.path.exists(saida_csv):
        save_header = False
    df.to_csv(saida_csv, mode='a', header=save_header, index=False)
    print('   Salvando arquivo', saida_csv, 'Ok')

    
    print(' Evolução - Início')
    #evolucao_carteira 3, remove ultimo registro informativo
    df = converte_tabela_evolucao(tabelas[3][:-1])
    
    # salvando arquivo
    save_header = True
    saida_csv = 'evolucao_carteira.csv'
    saida_csv = os.path.join('dados', saida_csv)
    if os.path.exists(saida_csv):
        save_header = False
    df.to_csv(saida_csv, mode='a', header=save_header, index=False)
    print('   Salvando arquivo', saida_csv, 'Ok')


    print(' Carteiras corretoras - Início')
    # se a página tiver as tabelas de corretoras
    # pega a primeira tabela de corretoras, para ter um dataframe inicial
    qtd_tabelas = len(tabelas)
    if qtd_tabelas >= 5:
        df = converte_tabela_corretora(tabelas[4][:-4])
    
    for i in range(5, qtd_tabelas):
        df2 = converte_tabela_corretora(tabelas[i][:-4])
        # percorre as tabelas com as corretoras
        df = pd.concat([df,df2.reindex(columns=df.columns)])
    
    # salvando arquivo
    save_header = True
    saida_csv = 'carteiras_corretoras.csv'
    saida_csv = os.path.join('dados', saida_csv)
    if os.path.exists(saida_csv):
        save_header = False
    df.to_csv(saida_csv, mode='a', header=save_header, index=False)
    print('   Salvando arquivo', saida_csv, 'Ok')

    
    print(' Carteiras corretoras - O que mudou - Início')
    # se a página tiver as tabelas de corretoras
    # pega a primeira tabela de corretoras, para ter um dataframe inicial
    qtd_tabelas = len(tabelas)
    if qtd_tabelas >= 5:
        df = converte_tabela_corretora_o_que_mudou(tabelas[4][-3:-1])
    
    for i in range(5, qtd_tabelas):
        df2 = converte_tabela_corretora_o_que_mudou(tabelas[i][-3:-1])
        # percorre as tabelas com as corretoras
        df = pd.concat([df,df2.reindex(columns=df.columns)])
    
    # salvando arquivo
    save_header = True
    saida_csv = 'carteiras_corretoras_o_que_mudou.csv'
    saida_csv = os.path.join('dados', saida_csv)
    if os.path.exists(saida_csv):
        save_header = False
    df.to_csv(saida_csv, mode='a', header=save_header, index=False)
    print('   Salvando arquivo', saida_csv, 'Ok')

print('*'*20)
print('Processo finalizado')
print('*'*20)
```
