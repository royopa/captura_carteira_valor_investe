```python
import pandas as pd
import os
import xlsxwriter
```

```python
# Cria o Excel de Saída
workbook = xlsxwriter.Workbook('saida.xlsx') 
```

```python
df = pd.read_csv(os.path.join('dados', 'evolucao_carteira.csv'))
df = df.sort_values(by=['corretora', 'ano_mes'], ascending=False)

df_resultado = pd.read_csv(os.path.join('dados', 'resultado_carteira.csv'))
df_resultado = df_resultado.sort_values(by=['nome', 'ano_mes'], ascending=False)
df_resultado

# The workbook object is then used to add new  
# worksheet via the add_worksheet() method. 
worksheet = workbook.add_worksheet('Resultado')

# Start from the first cell. 
row = 0
column = 1

for ano_mes in df['ano_mes'].unique().tolist():
    
    #################################
    ## Dados Resultado corretoras  ##
    #################################
    ano_mes_formatado = f'{str(ano_mes)[-2:]}/{str(ano_mes)[:4]}'
    filtro = df['ano_mes'] == ano_mes
    df_filtrado = df[filtro]
    df_filtrado = df_filtrado.sort_values(by=['corretora'], ascending=True)
    
    # preenche o título
    worksheet.write(0, column+1, ano_mes_formatado)

    row_detalhe=2
    # preenche o títulos
    worksheet.write(1, column, "Corretora")
    worksheet.write(1, column+1, "Mês %")
    worksheet.write(1, column+2, "Ano %")
    
    # percorre os registros de dados
    for index, line in df_filtrado.iterrows():
        column_ini = column

        # write operation perform 
        worksheet.write(row_detalhe, column_ini, line['corretora'])
        worksheet.write(row_detalhe, column_ini+1, line['desempenho_mes'])
        worksheet.write(row_detalhe, column_ini+2, line['desempenho_ano'])

        # incrementing the value of row by one 
        # with each iteratons
        row_detalhe += 1
        

    ###########################
    ## Dados Carteira Valor  ##
    ###########################
    row_detalhe += 1
    # preenche o título
    worksheet.write(row_detalhe, column_ini, 'CARTEIRA VALOR')
    
    # carteira valor
    nome = 'carteira_valor'
    ano_mes_formatado = f'{str(ano_mes)[-2:]}/{str(ano_mes)[:4]}'
        
    filtro = df_resultado['nome'] == nome
    df_filtrado = df_resultado[filtro]

    filtro = df_filtrado['ano_mes'] == ano_mes
    df_filtrado = df_filtrado[filtro]
        
    df_filtrado = df_filtrado.sort_values(by=['nome'], ascending=False)

    # percorre os registros de dados
    for index, line in df_filtrado.iterrows():
        column_ini = column

        worksheet.write(row_detalhe, column_ini+1, line['media_mes'])
        worksheet.write(row_detalhe+1, column_ini+1, line['acumulado_12_meses'])
        worksheet.write(row_detalhe, column_ini+2, line['acumulado_ano'])
            
        row_detalhe += 1
        worksheet.write(row_detalhe, column_ini, 'Acumulado 12 meses')
        
    
    ####################
    ## Dados Ibovespa ##
    ####################
    row_detalhe += 1
    row_detalhe += 1
    # preenche o título
    worksheet.write(row_detalhe, column_ini, 'IBOVESPA')
    
    # carteira valor
    nome = 'ibovespa'
    ano_mes_formatado = f'{str(ano_mes)[-2:]}/{str(ano_mes)[:4]}'
        
    filtro = df_resultado['nome'] == nome
    df_filtrado = df_resultado[filtro]

    filtro = df_filtrado['ano_mes'] == ano_mes
    df_filtrado = df_filtrado[filtro]
        
    df_filtrado = df_filtrado.sort_values(by=['nome'], ascending=False)

    #row_detalhe += 1
    #worksheet.write(row_detalhe+1, column_ini, 'IBOVESPA')
    #row_detalhe += 1    
    
    # percorre os registros de dados
    for index, line in df_filtrado.iterrows():
        worksheet.write(row_detalhe, column_ini+1, line['media_mes'])
        
        worksheet.write(row_detalhe+1, column_ini+1, line['acumulado_12_meses'])
        worksheet.write(row_detalhe, column_ini+2, line['acumulado_ano'])
        
        row_detalhe += 1
        worksheet.write(row_detalhe, column_ini, 'Acumulado 12 meses')        

    column += 3
```

```python
# The workbook object is then used to add new  
# worksheet via the add_worksheet() method. 
worksheet = workbook.add_worksheet('Carteira Valor')

df = pd.read_csv(os.path.join('dados', 'carteira_valor.csv'))
df = df.sort_values(by=['ano_mes', 'acao'], ascending=False)

# Start from the first cell. 
# Rows and columns are zero indexed. 
row = 0
column = 1

for ano_mes in df['ano_mes'].unique().tolist():
    ano_mes_formatado = f'{str(ano_mes)[-2:]}/{str(ano_mes)[:4]}'
    filtro = df['ano_mes'] == ano_mes
    df_filtrado = df[filtro]
    df_filtrado = df_filtrado.sort_values(by=['acao'], ascending=True)
    
    # escrevendo as colunas no excel, para cada grupo de meses
    # preenche o título
    worksheet.write(0, column, ano_mes_formatado)

    row_detalhe=2
    # preenche o títulos
    worksheet.write(1, column, "Ações")
    worksheet.write(1, column+1, "Desempenho")
    worksheet.write(1, column+2, "Indicações")
    
    # percorre os registros de dados
    for index, line in df_filtrado.iterrows():
        column_ini = column
        #print(line['acao'], line['desempenho'], line['indicacoes'])

        # write operation perform 
        worksheet.write(row_detalhe, column_ini, line['acao'])
        worksheet.write(row_detalhe, column_ini+1, line['desempenho'])
        worksheet.write(row_detalhe, column_ini+2, line['indicacoes'])

        # incrementing the value of row by one 
        # with each iteratons.
        row_detalhe += 1
    column += 4
```

```python
df = pd.read_csv(os.path.join('dados', 'carteiras_corretoras.csv'))
df = df.sort_values(by=['corretora', 'ano_mes'], ascending=False)
df

df_o_que_mudou = pd.read_csv(os.path.join('dados', 'carteiras_corretoras_o_que_mudou.csv'))
df_o_que_mudou = df_o_que_mudou.sort_values(by=['corretora', 'ano_mes'], ascending=False)
df_o_que_mudou

for corretora in sorted(df['corretora'].unique().tolist()):
    print(corretora)
    
    # cria a planillha corretora
    worksheet = workbook.add_worksheet(corretora)
    
    # Start from the first cell. 
    # Rows and columns are zero indexed. 
    row = 0
    column = 1
    
    # filtra apenas registros da corretora
    filtro = df['corretora'] == corretora
    df_filtrado = df[filtro]

    for ano_mes in df_filtrado['ano_mes'].unique().tolist():
        ano_mes_formatado = f'{str(ano_mes)[-2:]}/{str(ano_mes)[:4]}'
        filtro = df_filtrado['ano_mes'] == ano_mes
        df_filtrado2 = df_filtrado[filtro]
        df_filtrado2 = df_filtrado2.sort_values(by=['acao'], ascending=True)

        # escrevendo as colunas no excel, para cada grupo de meses
        # preenche o título
        titulo = f'{ano_mes_formatado}'
        worksheet.write(0, column, titulo)

        row_detalhe=2
        # preenche o títulos
        worksheet.write(1, column, "Ação")
        worksheet.write(1, column+1, "Variação")

        # percorre os registros de dados
        for index, line in df_filtrado2.iterrows():
            column_ini = column
            #print(line['acao'], line['desempenho'], line['indicacoes'])

            # write operation perform 
            worksheet.write(row_detalhe, column_ini, line['acao'])
            worksheet.write(row_detalhe, column_ini+1, line['variacao'])

            # incrementing the value of row by one 
            # with each iteratons.
            row_detalhe += 1

        #################
        ## O que mudou ##
        #################
            
        # filtra o que mudou 
        filtro = df_o_que_mudou['ano_mes'] == ano_mes 
        df_o_que_mudou2= df_o_que_mudou[filtro]
        
        filtro = df_o_que_mudou2['corretora'] == corretora
        df_o_que_mudou2= df_o_que_mudou2[filtro]
        
        df_o_que_mudou2 = df_o_que_mudou2.sort_values(by=['descricao'], ascending=True)        
        
        # subtítulo
        worksheet.write(row_detalhe, column_ini, 'O QUE MUDOU')
        row_detalhe += 1
        
        # percorre os registros de dados
        for index, line in df_o_que_mudou2.iterrows():
            worksheet.write(row_detalhe, column_ini, line['descricao'])
            # incrementing the value of row by one 
            # with each iteratons.
            row_detalhe += 1

        column += 3
```

```python
# Finally, close the Excel file 
# via the close() method. 
workbook.close() 
print('excel salvo com sucesso.')
```
